#### Imports

```python
import sys
sys.path.insert(0, "../..")
from shared.utils import chat
from IPython.display import Markdown, display
```

#### Define the steps 

```python
steps=["Generate a business problem","Identify pain points","Propose solutions","Bonus: prioritize solutions by impact"]
```

```python

resp_dict={}
for i,step in enumerate(steps):
    if i>0:
        prevStep=steps[i-1]
        prompt=f"{step} \n {resp_dict[prevStep]}"
    else:
        prompt=step
    resp_dict[step]=chat(step)
    display(Markdown(prompt))
    display(Markdown(resp_dict[step]))
    
          
    


   
```
